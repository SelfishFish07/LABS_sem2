```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from scipy.optimize import curve_fit
```

```python
def func(x, a):
    return a * x
```

```python
with open('Queue.csv', 'r') as file:
    reader = csv.reader(file)
    number=[]
    time_push=[]
    time_poll=[]
    for row in reader:
        number.append(int(row[0]))
        time_push.append(float(row[1]))
        time_poll.append(float(row[2]))
    time_push=np.array(time_push)
    time_poll=np.array(time_poll)
    number=np.array(number)
```

```python
plt.plot(number,time_push,marker='o',markersize=3,ls='', color='hotpink')
a, pcov = curve_fit(func, number, time_push)
plt.plot(number, a*number,lw=2,alpha=0.5,label=f"Аппроксимация $t={a[0]:.1e}n$")
plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Добавление элементов")
plt.savefig('../Images/queue_push.png')
plt.legend()
plt.show()
```

```python
plt.plot(number,time_poll,marker='o',markersize=3,ls='', color='hotpink')
a, pcov = curve_fit(func, number, time_poll)
plt.plot(number, a*number,lw=2,alpha=0.5,label=f"Аппроксимация $t={a[0]:.1e}n$")
plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Извлечение элементов")
plt.savefig('../Images/queue_poll.png')
plt.legend()
plt.show()
```
