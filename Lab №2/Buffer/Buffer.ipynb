```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from scipy.optimize import curve_fit
from PIL import Image
import os
```

```python
os.makedirs('../Images', exist_ok=True)
```

```python
def func(x, a):
    return a * x
```

```python
with open('Buffer.csv', 'r') as file:
    reader = csv.reader(file)
    number=[]
    timeA=[]
    timeB=[]
    timeC=[]
    for row in reader:
        number.append(int(row[0]))
        timeA.append(float(row[1]))
        timeB.append(float(row[2]))
        timeC.append(float(row[3]))
    number=np.array(number)
    timeA=np.array(timeA)
    timeB=np.array(timeB)
    timeC=np.array(timeC)
```

```python
plt.plot(number,timeA,marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, number, timeA)
plt.plot(number, func(number, *popt),lw=1,alpha=0.5,label=f"Аппроксимация $t={popt[0]:.1e}n$")
plt.xlabel("$n$, где $n$ - длина массива")
plt.ylabel("$t$, где $t$ - время в мкс")
plt.title("Стратегия А (+1)")
plt.savefig('../Images/A.png')
plt.legend()
plt.show()
```

```python
plt.plot(number,timeB,marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, number, timeB)
plt.plot(number, func(number, *popt),lw=1,alpha=0.5,label=f"Аппроксимация $t={popt[0]:.1e}n$")
plt.xlabel("$n$, где $n$ - длина массива")
plt.ylabel("$t$, где $t$ - время в мкс")
plt.title("Стратегия B (+100)")
plt.savefig('../Images/B.png')
plt.legend()
plt.show()
```

```python
def func2(x, a,b):
    return a * x+b
```

```python
plt.plot(number,timeC,marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func2, number, timeC)
plt.plot(number, func2(number, *popt),lw=1,alpha=0.5,label=f"Аппроксимация $t={popt[0]:.1e}n+{popt[1]:.1e}$")
plt.xlabel("$n$, где $n$ - длина массива")
plt.ylabel("$t$, где $t$ - время в мкс")
plt.title("Стратегия C (*2)")
plt.savefig('../Images/C.png')
plt.legend()
plt.show()
```
