```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from scipy.optimize import curve_fit
```

```python
def func(x, a,b):
    return a * x - b
```

```python
with open('Buffer.csv', 'r') as file:
    reader = csv.reader(file)
    number=[]
    timeA=[]
    timeB=[]
    timeC=[]
    for row in reader:
        number.append(int(row[0]))
        timeA.append(float(row[1]))
        timeB.append(float(row[2]))
        timeC.append(float(row[3]))
    timeA=np.array(timeA)
    timeB=np.array(timeB)
    timeC=np.array(timeC)
```

```python
plt.plot(np.log(number),np.log(timeA),marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, np.log(number), np.log(timeA))
plt.plot(np.log(number), func(np.log(number), *popt),lw=1,alpha=0.5,label=f"Аппроксимация $ln(t)={round(popt[0],2)}ln(n)+{round(popt[1],2)}$")
plt.xlabel("$ln(n)$, где $n$ - длина массива")
plt.ylabel("$ln(t)$, где $t$ - время в мкс")
plt.title("Стратегия А (+1)")
plt.legend()
plt.show()
```

```python
plt.plot(np.log(number),np.log(timeB),marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, np.log(number), np.log(timeB))
plt.plot(np.log(number), func(np.log(number), *popt),lw=1,alpha=0.5,label=f"Аппроксимация $ln(t)={round(popt[0],2)}ln(n)+{round(popt[1],2)}$")
plt.xlabel("$ln(n)$, где $n$ - длина массива")
plt.ylabel("$ln(t)$, где $t$ - время в мкс")
plt.title("Стратегия B (+100)")
plt.legend()
plt.show()
```

```python
plt.plot(np.log(number),np.log(timeC),marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, np.log(number), np.log(timeC))
plt.plot(np.log(number), func(np.log(number), *popt),lw=1,alpha=0.5,label=f"Аппроксимация $ln(t)={round(popt[0],2)}ln(n)+{round(popt[1],2)}$")
plt.xlabel("$ln(n)$, где $n$ - длина массива")
plt.ylabel("$ln(t)$, где $t$ - время в мкс")
plt.title("Стратегия C (*2)")
plt.legend()
plt.show()
```
