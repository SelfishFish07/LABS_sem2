```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from scipy.optimize import curve_fit
```

```python
def func1(x, a):
    return a * x**2

def func2(x, a):
    return a * x
```

```python
with open('Sum_mean.csv', 'r') as file:
    reader = csv.reader(file)
    number=[]
    time1=[]
    time2=[]
    for row in reader:
        number.append(int(row[0]))
        time1.append(float(row[1]))
        time2.append(float(row[2]))
```

```python
plt.plot(number,time1,marker='o',markersize=5,ls='', color='hotpink')
popt, pcov = curve_fit(func1, number, time1)
plt.plot(number, func1(np.array(number), *popt))
plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Поиск перебором. Произвольная сумма")
plt.show()
```

```python
plt.plot(number,time2,marker='o',markersize=5,ls='', color='hotpink')
popt, pcov = curve_fit(func2, number, time2)
plt.plot(number,func2(np.array(number), *popt))
plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Бинарный поиск. Произвольная сумма")
plt.show()
```
