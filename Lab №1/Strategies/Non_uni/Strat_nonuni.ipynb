```python
import matplotlib.pyplot as plt
import numpy as np
import csv
from scipy.optimize import curve_fit
```

```python
def func(x, a):
    return a * x
```

```python
with open('Strat_nonuni.csv', 'r') as file:
    reader = csv.reader(file)
    number=[]
    timeA=[]
    timeB=[]
    timeC=[]
    for row in reader:
        number.append(int(row[0]))
        timeA.append(float(row[1]))
        timeB.append(float(row[2]))
        timeC.append(float(row[3]))
```

```python
plt.plot(number,timeA,marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, number, timeA)
plt.plot(number, func(np.array(number), *popt),label=f"Аппроксимация $t={round(popt[0],3)}n$")
plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Стратегия А. Неравномерное распределение")
plt.legend()
plt.show()
```

```python
plt.plot(number,timeB,marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, number, timeB)
plt.plot(number, func(np.array(number), *popt),label=f"Аппроксимация $t={round(popt[0],3)}n$")
plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Стратегия B. Неравномерное распределение")
plt.legend()
plt.show()
```

```python
plt.plot(number,timeC,marker='o',markersize=4,ls='', color='hotpink')
popt, pcov = curve_fit(func, number, timeC)
plt.plot(number, func(np.array(number), *popt),label=f"Аппроксимация $t={popt[0]:.3f}n$")
plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Стратегия C. Неравномерное распределение")
plt.legend()
plt.show()
```
