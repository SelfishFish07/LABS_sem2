```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
with open('Search_mean.csv', 'r') as file:
    reader = csv.reader(file)
    number=[]
    time1=[]
    time2=[]
    for row in reader:
        number.append(int(row[0]))
        time1.append(float(row[1]))
        time2.append(float(row[2]))
```

```python
plt.plot(number,time1,marker='o',markersize=5,ls='', color='hotpink',label='Поиск перебором',alpha=0.7)
plt.plot(number,time2,marker='o',markersize=5,ls='', color='slateblue',label='Бинарный поиск',alpha=0.7)

plt.xlabel("Длина массива $n$")
plt.ylabel("Время $t$, мкс")
plt.title("Поиск произвольного элемента массива")
plt.legend()
plt.show()
```
